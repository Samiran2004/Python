```python
def printRectangleStarPattern(n, m):
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            print("*", end=" ")
        print()
```

```python
n: int = int(input("Enter the value of n: "))
m: int = int(input("Enter the value of m : "))
printRectangleStarPattern(n, m)
```
