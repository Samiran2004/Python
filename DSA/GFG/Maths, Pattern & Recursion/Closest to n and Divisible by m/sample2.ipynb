```python
def closest_number(n, m):
    q = int(n / m)

    # First possible
    n1 = m * q

    # Second possible
    if (n * m) > 0:
        n2 = (m * (q + 1))
    else:
        n2 = (m * (q - 1))

    if abs(n - n1) < abs(n - n2):
        return n1
    return n2
```

```python
n: int = int(input("Enter the value of n: "))
m: int = int(input("Enter the value of m: "))

print(closest_number(n, m))
```
