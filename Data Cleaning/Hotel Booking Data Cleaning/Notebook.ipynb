```python
import pandas as pd
```

```python
data = pd.read_csv('hotel_bookings.csv')
data
```

```python
#drop/rename columns

data.columns
```

```python
rename_column_list = {
    'adults': 'num_of_adults',
    'children': 'num_of_children',
    'babies': 'num_of_babies'
}

data.rename(rename_column_list, axis=1, inplace=True)

data.columns
```

```python
# Dealing with NaNs
data.isnull().sum() * 100 / len(data)
```

```python
data['agent'].fillna(-1, inplace=True)
```

```python
data.isnull().sum() * 100 / len(data)
```

```python
data[data['num_of_children'].isna()]
```

```python
data['num_of_children'].fillna(0, inplace=True)
```

```python
data[data['country'].isna()]
```

```python
data[data['country'].isna()][['hotel', 'is_canceled', 'country']]
```

```python
data['country'].fillna('Unknown', inplace=True)
```

```python
data.dropna(subset=['num_of_children'], inplace=True)
```

```python
data.drop(columns=['company'], inplace=True)
```

```python
data.isnull().sum() * 100 / len(data)
```

```python
# Check column's datatype

data.dtypes
```

```python
data = data.astype({
    'is_canceled': 'boolean',
    'is_repeated_guest': 'boolean',
    'num_of_children': 'int'
})

data.dtypes
```

```python
data['lead_time'].describe()
```

```python
# Bin
bins = [0, 100, 200, 300, 400, 500, 600, 700, 800, 900]
labels = ['0-100', '101-200', '201-300', '301-400', '401-500', '501-600', '601-700', '701-800', '801-900']
data['lead_time_binned'] = pd.cut(data['lead_time'], bins=bins, labels=labels)

data[['lead_time', 'lead_time_binned']]
```

```python
# Seperate columns

data.head()
```

```python
data['reservation_status_year'] = data['reservation_status_date'].str.split('-', expand=True)[0]

data.head()
```

```python
data['hotel'].unique()

# String cleaning...
data['hotel'] = data['hotel'].replace(r"[\*\n\^]", '', regex=True)

data['hotel'].unique()
```

```python
# Remove duplicates

data.duplicated(keep=False)
```

```python
data.loc[data.duplicated(keep=False)]
```

```python
data.drop_duplicates(keep='first', inplace=True)
data.loc[data.duplicated(keep=False)]
```
