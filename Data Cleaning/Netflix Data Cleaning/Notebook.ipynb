```python
# Import dataset...

data = pd.read_csv("data/datasets/ariyoomotade/netflix-data-cleaning-analysis-and-visualization/versions/1/netflix1.csv")

data
```

```python
cleaned_data = data.copy()

cleaned_data
```

```python
cleaned_data.info()
```

### Deleting redundant columns...

```python
cleaned_data.columns
```

```python
cleaned_data[["rating"]]
```

```python
cleaned_data.drop(columns=["rating"], inplace=True)
```

```python
cleaned_data.columns
```

### Renaming columns...

```python
new_column_names = []

for i in cleaned_data.columns:
    new_column_names.append(i.capitalize())

cleaned_data.columns = new_column_names

cleaned_data.columns
```

### Dropping duplicated rows...

```python
cleaned_data.duplicated().sum()
```

```python
cleaned_data.drop_duplicates(keep='first', inplace=True)
```

### Drop Nan...

```python
cleaned_data.isna().sum()
```

```python
cleaned_data.dropna(axis=0, inplace=True)
```

### Check for some more transformation...

```python
cleaned_data['Show_id'].unique()
```

```python
# Extract the numeric part after 's' in Show_id
cleaned_data["Show_id_numeric"] = cleaned_data["Show_id"].str[1:]

cleaned_data["Show_id"] = cleaned_data["Show_id_numeric"].astype(int)

cleaned_data["Show_id"]
```

```python
cleaned_data["Date_added"]
```

```python
cleaned_data["Date_added"] = cleaned_data["Date_added"].apply(lambda x: x.replace("/", "-"))

cleaned_data["Date_added"] = cleaned_data["Date_added"].astype(str)

cleaned_data["Date_added"]
```

```python
cleaned_data.Country.value_counts()
```

```python
cleaned_data
```

```python
print(f"Total null values: {cleaned_data.isna().sum()}")

print(f"Total duplicate values: {cleaned_data.duplicated().sum()}")
```
