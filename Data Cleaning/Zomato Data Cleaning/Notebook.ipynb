```python
cleaned_data = data.copy()

cleaned_data.columns
```

```python
# Deleting redundant columns

columns_to_keep = ['name', 'online_order', 'book_table', 'rate', 'dish_liked', 'approx_cost(for two people)']

columns_to_drop = ['url', 'address', 'votes','phone', 'location', 'rest_type', 'cuisines', 'reviews_list',
                    'menu_item','listed_in(type)', 'listed_in(city)']
```

```python
print(f"Length of keep columns: {len(columns_to_keep)}")
print(f"Length of columns to drop: {len(columns_to_drop)}")
```

## 2nd way -> Filtering Data

```python
cleaned_data.drop(columns_to_drop, inplace=True, axis=1)

cleaned_data.columns
```

```python
# Renaming data

for i in data.columns:
    print(i.capitalize())
```

```python
new_column_name = []

for i in cleaned_data.columns:
    new_column_name.append(i.capitalize())

cleaned_data.columns = new_column_name

cleaned_data.columns
```

### Dropping duplicates...

```python
data.duplicated().value_counts()
```

```python
# cleaned_data.duplicated().sum()
cleaned_data.drop_duplicates(keep='first', inplace=True)
cleaned_data.duplicated().sum()
```

```python
cleaned_data.duplicated().value_counts()
```

```python
# Removing NaN value

cleaned_data.isna().sum()
```

```python
cleaned_data.dropna(subset=["Dish_liked"], inplace=True)

cleaned_data.isna().sum()
```

```python
cleaned_data.dropna(inplace=True)
```

```python
cleaned_data.reset_index(inplace=True)

cleaned_data
```

```python
cleaned_data.drop(columns="index", inplace=True)

cleaned_data
```

```python
cleaned_data["Rate"] = cleaned_data["Rate"].str.replace(" ", "")

cleaned_data["Rate"]=cleaned_data["Rate"].str.replace("/", " out of ")

cleaned_data["Rate"]
```

```python
cleaned_data["Online_order"] = cleaned_data["Online_order"].apply(lambda x: 1 if x == "Yes" else 0)
```

```python
cleaned_data.to_csv("./data/zomato_cleaned_data.csv")
```
