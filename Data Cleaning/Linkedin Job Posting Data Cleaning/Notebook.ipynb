```python
posting_data = pd.read_csv('datas/postings.csv')

posting_data
```

```python
posting_cleaned_dataset = posting_data.copy()
posting_cleaned_dataset
```

```python
posting_cleaned_dataset.columns
```

```python
# Change the title to job_title, applies to nums_applicants, description to job_description

posting_cleaned_dataset.rename({
    'title': 'job_title',
    'applies': 'nums_applicants',
    'description': 'job_description'
}, axis=1, inplace=True)

posting_cleaned_dataset.head()
```

```python
posting_cleaned_dataset.dtypes
```

```python
posting_cleaned_dataset = posting_cleaned_dataset.astype({
    'zip_code': 'str',
    "fips": 'str'
})

posting_cleaned_dataset.dtypes
```

```python
posting_data.info()
```

```python
posting_cleaned_dataset.drop(['med_salary', 'closed_time', 'skills_desc'], axis=1, inplace=True)

posting_cleaned_dataset.columns
```

### String cleaning

```python
posting_cleaned_dataset['job_title'] = posting_cleaned_dataset['job_title'].str.strip()
```

```python
posting_cleaned_dataset['job_title'] = posting_cleaned_dataset['job_title'].str.lower()

posting_cleaned_dataset['job_title']
```

### Handle missing values

```python
posting_cleaned_dataset.isna().sum()
```

```python
posting_cleaned_dataset.isnull().sum() * 100 / len(posting_cleaned_dataset.isnull())
```

```python
posting_cleaned_dataset.info()
```

```python
posting_cleaned_dataset['remote_allowed'].unique()
```

```python
posting_cleaned_dataset['remote_allowed'].fillna(0, inplace=True)
```

```python
posting_cleaned_dataset['remote_allowed'].unique()
```

```python
posting_cleaned_dataset['remote_allowed'] = posting_cleaned_dataset['remote_allowed'].astype('int')

posting_cleaned_dataset['remote_allowed'].unique()
```

```python
# Drop rows...
posting_cleaned_dataset.dropna(axis=0, how='all', inplace=True)
```

```python
posting_cleaned_dataset.shape
```

```python
posting_data.shape   # Original dataset
```

```python
posting_cleaned_dataset.dropna(axis=0, subset=['company_name'], inplace=True)

posting_cleaned_dataset.shape
```

### Checking for duplicated values

```python
posting_cleaned_dataset = pd.concat([posting_cleaned_dataset, posting_cleaned_dataset.iloc[0:2]], ignore_index=True)

posting_cleaned_dataset = pd.concat([posting_cleaned_dataset, posting_cleaned_dataset.iloc[1000: 1002]], ignore_index=True)

posting_cleaned_dataset = pd.concat([posting_cleaned_dataset, posting_cleaned_dataset.iloc[10000: 10000]], ignore_index=True)
```

```python
posting_cleaned_dataset.duplicated()
```

```python
posting_cleaned_dataset[posting_cleaned_dataset.duplicated()]
```

```python
posting_cleaned_dataset.drop_duplicates(inplace=True)

posting_cleaned_dataset[posting_cleaned_dataset.duplicated()]
```

```python
posting_cleaned_dataset.isnull().sum() * 100 / len(posting_cleaned_dataset.isnull())
```

```python
posting_cleaned_dataset.dropna(subset=['normalized_salary'], inplace=True, axis=0)

posting_cleaned_dataset.isnull().sum() * 100 / len(posting_cleaned_dataset.isnull())
```

```python
posting_cleaned_dataset['max_salary']
```

```python
posting_cleaned_dataset['max_salary'].fillna(0, inplace=True)
posting_cleaned_dataset['max_salary'].isna().sum()
```

```python
posting_cleaned_dataset.isnull().sum() * 100 / len(posting_cleaned_dataset.isnull())
```

```python
posting_cleaned_dataset["views"]
```

```python
posting_cleaned_dataset['views'].fillna(0, inplace=True)

posting_cleaned_dataset['views'] = posting_cleaned_dataset['views'].astype('int')

posting_cleaned_dataset['views']
```

```python
# Populate min_salary

posting_cleaned_dataset['min_salary'].fillna(0.0, inplace=True)

# Populate nums_applicants

posting_cleaned_dataset['nums_applicants'].fillna(0.0, inplace=True)

# Populate application_url

posting_cleaned_dataset.dropna(subset=['application_url'], inplace=True)

# Populate formatted_experience_level
posting_cleaned_dataset.drop('formatted_experience_level', axis=1, inplace=True)
```

```python
# Populate posting_domain

posting_cleaned_dataset['posting_domain'] = posting_cleaned_dataset['posting_domain'].str.strip()

posting_cleaned_dataset['posting_domain'] = posting_cleaned_dataset['posting_domain'].str.replace(r"\n\^\.", "", regex=True)

posting_cleaned_dataset['posting_domain'].fillna("Unknown", inplace=True)
```

```python
posting_cleaned_dataset.isnull().sum() * 100 / len(posting_cleaned_dataset.isnull())
```

```python
os.makedirs('cleaned_data', exist_ok=True)

posting_cleaned_dataset.to_csv('./cleaned_data/postings_cleaned_data.csv')
```
